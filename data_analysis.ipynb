# Dataset analysis

```python
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
base_path = "data/perphect-data"
```

### Public dataset

```python
public_couples_df = pd.read_csv(os.path.join(base_path, "public_data_set", "couples_df.csv"))
public_phages_df = pd.read_csv(os.path.join(base_path, "public_data_set", "phages_df.csv"))
public_bacteria_df = pd.read_csv(os.path.join(base_path, "public_data_set", "bacteria_df.csv"))
print('public_couples_df.shape = ', public_couples_df.shape)
print('public_phages_df.shape = ', public_phages_df.shape)
print('public_bacteria_df.shape = ', public_bacteria_df.shape)
```

#### Sequence length

Bacteria sequence length 

```python
plt.figure(figsize=(8, 5))
plt.hist(public_bacteria_df['sequence_length'], bins=50)
plt.xlabel('Bacteria Sequence Length')
plt.ylabel('Count')
plt.title('Distribution of Bacteria Sequence Length (Public Dataset)')
plt.show()
```

On PERPHECT paper, they used 7M as the maximum length for the bacteria sequences.

### Joined datasets

```python
merged_bacteria = pd.merge(public_bacteria_df, private_bacteria_df, on='bacterium_id', suffixes=('_public', '_private'))

# Check if sequences are identical for matching ids
merged_bacteria['same_sequence'] = merged_bacteria['bacterium_sequence_public'] == merged_bacteria['bacterium_sequence_private']

merged_bacteria[['bacterium_id', 'same_sequence']]
```

```python
merged_phages = pd.merge(public_phages_df, private_phages_df, on='phage_id', suffixes=('_public', '_private'))

# Check if sequences are identical for matching ids
merged_phages['same_sequence'] = merged_phages['phage_sequence_public'] == merged_phages['phage_sequence_private']

merged_phages[['phage_id', 'same_sequence']]
```

Each ID corresponds to a different individual in the combined dataset

### Undersample private dataset

```python
from imblearn.under_sampling import RandomUnderSampler
```

```python
rus = RandomUnderSampler(random_state=42)
X = private_couples_df[['phage_id', 'bacterium_id']]
y = private_couples_df['interaction_type']
X_res, y_res = rus.fit_resample(X, y) # type: ignore
X_res['interaction_type'] = y_res
X_res
```

```python
X_res['interaction_type'].value_counts()
```

```python
# Re add the 'id' column from the original private_couples_df
X_res = X_res.merge(
    private_couples_df[['id', 'phage_id', 'bacterium_id', 'interaction_type']],
    on=['phage_id', 'bacterium_id', 'interaction_type'],
    how='left'
)

# Combine it with public couples
combined_couples_df = pd.concat([X_res, public_couples_df], ignore_index=True)
combined_couples_df
```

```python
combined_couples_df["interaction_type"].value_counts()
```

```python
os.makedirs(os.path.join(base_path, "all-private-undersampled"), exist_ok=True)
combined_couples_df.to_csv(os.path.join(base_path, "all-private-undersampled/couples_df.csv"), index=False, columns=['id', 'phage_id', 'bacterium_id', 'interaction_type'])
```

### Oversample private dataset

```python
from imblearn.over_sampling import RandomOverSampler

ros = RandomOverSampler(random_state=42)
X = private_couples_df[['phage_id', 'bacterium_id']]
y = private_couples_df['interaction_type']
X_res, y_res = ros.fit_resample(X, y) # type: ignore
X_res['interaction_type'] = y_res

# Re add the 'id' column from the original private_couples_df
X_res = X_res.merge(
    private_couples_df[['id', 'phage_id', 'bacterium_id', 'interaction_type']],
    on=['phage_id', 'bacterium_id', 'interaction_type'],
    how='left'
)

# Combine it with public couples
combined_couples_df = pd.concat([X_res, public_couples_df], ignore_index=True)
combined_couples_df
```

```python
combined_couples_df["interaction_type"].value_counts()
```

```python
os.makedirs(os.path.join(base_path, "all-private-oversampled"), exist_ok=True)
combined_couples_df.to_csv(os.path.join(base_path, "all-private-oversampled/couples_df.csv"), index=False, columns=['id', 'phage_id', 'bacterium_id', 'interaction_type'])
```
